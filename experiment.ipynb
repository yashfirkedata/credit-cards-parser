```python
# """
# Extract selected mails from your gmail account

# 1. Make sure you enable IMAP in your gmail settings
# (Log on to your Gmail account and go to Settings, See All Settings, and select
#  Forwarding and POP/IMAP tab. In the "IMAP access" section, select Enable IMAP.)

# 2. If you have 2-factor authentication, gmail requires you to create an application
# specific password that you need to use.
# Go to your Google account settings and click on 'Security'.
# Scroll down to App Passwords under 2 step verification.
# Select Mail under Select App. and Other under Select Device. (Give a name, e.g., python)
# The system gives you a password that you need to use to authenticate from python.

# """

# Importing libraries
import imaplib
import email
import yaml  #To load saved login credentials from a yaml file

# with open("credentials.yml") as f:
#     content = f.read()

# # from credentials.yml import user name and password
# my_credentials = yaml.load(content, Loader=yaml.FullLoader)

# #Load the user name and passwd from yaml file
# user, password = my_credentials["user"], my_credentials["password"]

user = "yash.subs.ae.2@gmail.com"
password = "bcfv ujdk xwad hrog" # Please ensure this is an app-specific password if 2FA is enabled
#URL for IMAP connection
imap_url = 'imap.gmail.com'

# Connection with GMAIL using SSL
my_mail = imaplib.IMAP4_SSL(imap_url)

# Log in using your credentials
try:
    my_mail.login(user, password)
    print("Successfully logged in.")
except imaplib.IMAP4.error as e:
    print(f"Login failed: {e}")
    print("Please ensure your username and password are correct, and that IMAP is enabled.")
    print("If you use 2-Factor Authentication, make sure to use an App Password.")
    exit()

# Select the Inbox to fetch messages
status, messages = my_mail.select('Inbox')
if status != 'OK':
    print(f"Failed to select Inbox: {messages}")
    exit()
print(f"Inbox selected. Total messages: {messages[0].decode('utf-8')}")


# Define the search term
search_term = "Credit Card"

# Define the IMAP search criterion
# This will search for the term in the email's subject OR body.
# The term is enclosed in double quotes as per IMAP standards, especially if it contains spaces.
# For other keys (criteria): https://gist.github.com/martinrusev/6121028#file-imap-search
criterion = f'(OR (SUBJECT "{search_term}") (BODY "{search_term}"))'
# Alternative: search in all text parts of the email (including headers and body).
# This might be broader and could also work.
# criterion = f'TEXT "{search_term}"'

print(f"Searching for emails with criterion: {criterion}")
# _, data = my_mail.search(None, key, value)  # Original search
typ, data = my_mail.search(None, criterion)  # Search for emails based on the new criterion

if typ != 'OK':
    print(f"Error searching for emails: {data[0].decode('utf-8') if data else 'No data'}")
    my_mail.logout()
    exit()

mail_id_list = data[0].split()  #IDs of all emails that we want to fetch
print(f"Found {len(mail_id_list)} emails matching the criteria.")

msgs = [] # empty list to capture all messages
#Iterate through messages and extract data into the msgs list
for num in mail_id_list:
    # Fetch the email message by ID
    # RFC822 returns the entire message (headers and body)
    typ, fetch_data = my_mail.fetch(num, '(RFC822)')
    if typ == 'OK':
        msgs.append(fetch_data)
    else:
        print(f"Error fetching email ID {num.decode('utf-8')}")


#Now we have all messages, but with a lot of details
#Let us extract the right text and print on the screen

#In a multipart e-mail, email.message.Message.get_payload() returns a
# list with one item for each part. The easiest way is to walk the message
# and get the payload on each part:
# https://stackoverflow.com/questions/1463074/how-can-i-get-an-email-messages-text-content-using-python

# NOTE that a Message object consists of headers and payloads.
print(f"\n--- Displaying Extracted Emails ({len(msgs)} found) ---")
if not msgs:
    print("No emails found matching your criteria or an error occurred during fetch.")

for i, msg_data in enumerate(msgs[::-1]): # Process newest first
    print(f"\n--- Email {i+1} ---")
    for response_part in msg_data:
        if isinstance(response_part, tuple):
            try:
                my_msg = email.message_from_bytes(response_part[1])
                print("_________________________________________")
                print("Subject:", my_msg['subject'])
                print("From:", my_msg['from'])
                # print ("To:", my_msg['to']) # You can also print To, Date, etc.
                # print ("Date:", my_msg['date'])
                print("Body:")

                body_content = ""
                if my_msg.is_multipart():
                    for part in my_msg.walk():
                        content_type = part.get_content_type()
                        content_disposition = str(part.get("Content-Disposition"))

                        # Skip attachments
                        if "attachment" not in content_disposition:
                            if content_type == 'text/plain':
                                try:
                                    body_content += part.get_payload(decode=True).decode('utf-8', errors='replace') + "\n"
                                except Exception as e:
                                    body_content += f"[Could not decode part: {e}]\n"
                            elif content_type == 'text/html':
                                # You could add HTML parsing here if needed, or just indicate its presence
                                # For now, we are primarily interested in plain text.
                                pass # Or, try to extract text from HTML if you want
                else: # Not a multipart email, just a single body
                    if my_msg.get_content_type() == 'text/plain':
                        try:
                            body_content = my_msg.get_payload(decode=True).decode('utf-8', errors='replace')
                        except Exception as e:
                            body_content = f"[Could not decode payload: {e}]"

                if body_content.strip():
                    print(body_content.strip())
                else:
                    print("[No plain text body found or body is empty]")
            except Exception as e:
                print(f"[Error parsing email: {e}]")

# Logout from the server
my_mail.logout()
print("\nLogged out.")
```
